```python
# importing the libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading data from the flights dataset
# source https://github.com/mwaskom/seaborn-data/blob/master/flights.csv
df = pd.read_csv('flights.csv')
```

```python
# Peeking into the dataset
df.head()
```

```python
# Stastistical analysis of number passenger over time
data = df[['year', 'passengers']].groupby(by = ['year']).sum().reset_index().sort_values(by='year',ascending=True)
```

```python
# Calculating basic statistical analysis on the passengers
data['passengers'].describe().reset_index()
```

```python
# Plotting Time(in year) vs Total Count graph
f,ax=plt.subplots(figsize=(15,10))
sns.barplot(x='year', y='passengers', data=data, color='g')
ax.set(xlabel='Year', ylabel='Total Number of Passengers')
plt.show()
```

```python
# Getting total flight month wise
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data = df[['month','passengers']].groupby('month').sum().reindex(month_order, axis=0).reset_index()
```

```python
# Calculating basic statistical analysis on the passengers
data['passengers'].describe().reset_index()
```

```python
# Plotting Time(in month) vs Total Count graph
f,ax=plt.subplots(figsize=(15,10))
sns.barplot(x='month', y='passengers', data=data, color='g')
ax.set(xlabel='Month', ylabel='Total Number of Passengers')
plt.show()
```

```python
# Reading data from the iris dataset
# source https://github.com/mwaskom/seaborn-data/blob/master/iris.csv
df = pd.read_csv('iris.csv')
```

```python
# Plotting sepal_length vs petal_length for iris flowers

f,ax=plt.subplots(figsize=(15,10))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
ax.set(xlabel='Sepal Length', ylabel='Petal Length')
plt.show()
```

```python
# Plotting sepal_width vs petal_width for iris flowers

f,ax=plt.subplots(figsize=(15,10))
sns.scatterplot(data=df, x="sepal_width", y="petal_width", hue="species")
ax.set(xlabel='Sepal Width', ylabel='Petal Width')
plt.show()
```
